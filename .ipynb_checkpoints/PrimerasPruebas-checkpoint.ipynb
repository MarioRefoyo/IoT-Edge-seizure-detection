```python
from pylab import *
from scipy import signal
import numpy as np
import scipy.fftpack as fourier
from scipy.signal import butter, filtfilt, lfilter
```

```python
data = loadtxt("prueba_1000Hz.txt")
fs=1000
```

```python
x = data[:, 5]
```

```python
def highpass(s, cutoff, order = 5, fs):
    b, a = signal.butter(order, cutoff / (fs/2), btype='high', analog=False)
    return filtfilt(b, a, s)

def lowpass(s, f, order=10, fs):
    b, a = signal.butter(order, f / (fs/2))
    return lfilter(b, a, s)

def notchfilter(s, f, Q = 12, fs):
    b, a = signal.iirnotch(f / (fs/2), Q)
    return lfilter(b, a, s)
```

```python
xpa = highpass(x, 0.5)
```

```python
xpb = lowpass(xpa, 49)
```

```python
xf = notchfilter(xpb, 50)
```

```python
 f, t, Zxx = signal.stft(xf, fs, nperseg=1000)
```

```python
shape(Zxx[0:50])
```

```python
pcolormesh(t, f[0:50], np.abs(Zxx)[0:50], vmin=0, vmax=60)
```

```python
t = linspace(0, size(x)/fs, size(x))
plot(t, x)
```

```python
size(x)
```

```python
xorg = reshape(x[0:184000], (1000, 184))
```

```python
shape(xorg[:, 120])
```

```python
xex = reshape(xorg[:, 160:164], (4000,))
```

```python
shape(xex)
```

```python
shape(xorg[100])
```

```python
win = 4*fs
freqs, psd = signal.welch(xex, fs, nperseg=win)
```

```python
plot (freqs[0:200], psd[0:200])
```

```python
def highpass(s, cutoff, order = 5, fs = 100.0):
    b, a = signal.butter(order, cutoff / (fs/2), btype='high', analog=False)
    return filtfilt(b, a, s)

def lowpass(s, f, order=10, fs= 100.0):
    b, a = signal.butter(order, f / (fs/2))
    return lfilter(b, a, s)

def notchfilter(s, f, Q = 12, fs= 100.0):
    b, a = signal.iirnotch(f / (fs/2), Q)
    return lfilter(b, a, s)
```

```python
data = loadtxt("ojos_cerrados_y_blinks_aluminio.txt")
x = data[:, 5]
xpa = highpass(x, 0.5)
xpb = lowpass(xpa, 49)
xf = notchfilter(xpb, 50)
f, t, Zxx = signal.stft(xf, fs, nperseg=100)
shape(Zxx[0:50])
pcolormesh(t, f[0:50], np.abs(Zxx)[0:50], vmin=0, vmax=60)
```

```python
t = linspace(0, size(x)/fs, size(x))
plot(t, x)
```

```python
size(x)
```

```python
xorg = reshape(x[0:15000], (100, 150))
xex = reshape(xorg[:, 110:114], (400,))
win = 4*fs
freqs, psd = signal.welch(xex, fs, nperseg=win)
plot (freqs[0:200], psd[0:200])
```
