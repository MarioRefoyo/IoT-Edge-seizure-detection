```python
def lowpass(s, f, order=4, fs=100.0):
    b, a = signal.butter(order, f / (fs/2))
    return signal.lfilter(b, a, s)
```

```python
def butter_highpass(cutoff, fs=100.0, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
    return b, a

def highpass(data, cutoff, fs = 100.0, order=5):
    b, a = butter_highpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y
```

```python
plot(highpass(data[:,5][3000:4000],1))
```

```python
plot(lowpass(data[:,5], 45))
```
