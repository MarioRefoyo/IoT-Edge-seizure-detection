```python
from pyedflib import highlevel
import numpy as np
from pylab import *
import csv
from scipy import signal
import pandas as pd
import seaborn
```

```python
class Register:
    
    def __init__(self, name, fs, nseizures):
        self.name = name
        self.fs = fs
        self.nseizures = nseizures
        self.seizures = []
            
    def addSeizure (self, start, end):
        seizure = (start, end)
        self.seizures.append(seizure)
```

```python
def frec2sample_range(fi, fo, fs, N):
    si = max(1,floor(fi*N/fs))
    so = ceil(fo*N/fs)
    return int(si), int(so)

def band_energy(fft, fs):
    N = len(fft)
    # Energia total
    et = sum(fft)
    dsi, dso = frec2sample_range(0.5, 4, fs, N)
    d = sum(fft[dsi:dso])
    tsi, tso = frec2sample_range(4, 7, fs, N)
    t = sum(fft[tsi:tso])
    asi, aso = frec2sample_range(7, 13, fs, N)
    a = sum(fft[asi:aso])
    bsi, bso = frec2sample_range(13, 30, fs, N)
    b = sum(fft[bsi:bso])
    gsi, gso = frec2sample_range(30, 50, fs, N)
    g = sum(fft[gsi:gso])
    return et, d, t, a, b, g

def chunk_processing(chunk, fs):
    
    ninstances = chunk.shape[0]
    et = zeros(ninstances)
    d = zeros(ninstances)
    t = zeros(ninstances)
    a = zeros(ninstances)
    b = zeros(ninstances)
    g = zeros(ninstances)
    
    spectral_centroid = zeros(ninstances)
    variatonial_coeff = zeros(ninstances)
    spectral_skew = zeros(ninstances)
    
    #df = pd.DataFrame()
    features = ['total_energy', 'delta', 'theta', 'alpha', 'beta', 'gamma', 'spectral_centroid', 'variatonial_coeff', 'spectral_skew']
    
    indexStart = chunk.index[0]
    for index, row in chunk.iterrows():
        index = index - indexStart
        signal = row[0:512].values
        X = fft(signal)
        Px = real(X*conj(X))
        et[index], d[index], t[index], a[index], b[index], g[index] = band_energy(Px, fs)
         
        p_spectrum = Px[:len(Px)//2]/sum(Px[:len(Px)//2])
        normalized_f = linspace(0, 1, len(Px)//2)
        spectral_centroid[index] = sum(normalized_f * p_spectrum)
        variatonial_coeff[index] = sum( (((normalized_f - spectral_centroid[n]))**2) * p_spectrum) / sum(p_spectrum)
        spectral_skew[index] = sum( ((normalized_f - spectral_centroid[n])/variatonial_coeff[n])**3 * p_spectrum)

        
    data = [et, d, t, a, b, g, spectral_centroid, variatonial_coeff, spectral_skew]
    data = np.array(data).transpose()
    df = pd.DataFrame(data, columns = features)
    df['seizure'] = pd.Series(chunk['seizure'].values, index = df.index)
        
    return df
```

```python
import glob, os

rootdir = "E:\Master\TFM\chb-mit-scalp-eeg-database-1.0.0\chb01"
```

```python
os.chdir(rootdir)
annotation = glob.glob('*txt')

with open(annotation[0]) as f:
    registers = {}
    for line in f:
        if ("Data Sampling Rate" in line):
            line = line.split()
            fs = int(line[3])
        elif ("File Name" in line):
            name = line.split()[2]
            while True:
                newLine = f.readline()
                if ("Number of Seizures" in newLine):
                    nseizures = int(newLine.split()[5])
                    register = Register(name, fs, nseizures)
                    if nseizures > 0:
                        for i in range(nseizures):
                            line1 = f.readline().split()
                            line2 = f.readline().split()
                            if (line1[3] == "Time:"):
                                start = int(line1[4])
                                end = int(line2[4])
                            else:
                                start = int(line1[3])
                                end = int(line2[3])
                            register.addSeizure(start, end)
                            
                    registers[name] = register
                    break
```

# Ejemplo para un solo canal

```python
timeW = 2
N = timeW*fs

registro = 3
ncanales = 1
example = pd.DataFrame()

index = 0
for key, value in registers.items():
    df = pd.DataFrame()
    if index==registro-1:
        # Signal reading
        signals, signal_headers, header = highlevel.read_edf(key)
        signals = np.float16(signals)
        # Truncate to obtain windows
        samples = signals.shape[1]
        nw = int(samples//N)
        signals = signals[:, 0:nw*N]
        samples = signals.shape[1]
        time = np.linspace(0, signals.shape[1]/fs, signals.shape[1])
        
        print("Readed " + key)

        # Seizure vector creation
        seizure = zeros(samples)
        if (len(value.seizures) > 0):
            for n in range (len(value.seizures)):
                start = value.seizures[n][0]*fs
                end = value.seizures[n][1]*fs
                seizure[start:end] = np.ones(end-start)

        seizureW = np.reshape(seizure, [nw, N])
        seizureW = (sum(seizureW, 1) > N//2)
        
        # plot the register
        plt.figure(figsize=(50,150))
        nS = signals.shape[0]
        for n, s in enumerate(signals):
            plt.subplot(nS, 1, n+1)
            plt.plot(time, s)
            plt.plot(time, seizure*np.amax(signals), linewidth=4)
        
        

        # Create list for feature names
        features = []
        for i in range(N):
            features.append("sample" + str(i+1))

        # Create register dataframe
        #df = pd.DataFrame()
        for channel, signal in enumerate(signals):
            if channel>ncanales-1:
                break
            newSignal = np.reshape(signal, [nw, N])
            newdf = pd.DataFrame(newSignal, columns = features)
            newdf['seizure'] = pd.Series( seizureW, index = newdf.index)
            df = df.append(newdf, ignore_index=True)
            print("Created rows for " + str(channel) + " channel")

    index += 1

    # Add to the patient dataframe
    example = example.append(df, ignore_index=True)
    
```
