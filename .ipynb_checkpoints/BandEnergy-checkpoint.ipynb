Preprocess the signal


```python
from pylab import *
from scipy import signal
import numpy as np
import scipy.fftpack as fourier
from scipy.signal import butter, filtfilt, lfilter

from PyEMD import EMD
from PyEMD import EEMD

def highpass(s, cutoff, fs, order = 5):
    b, a = signal.butter(order, cutoff / (fs/2), btype='high', analog=False)
    return filtfilt(b, a, s)

def lowpass(s, f, fs, order=10):
    b, a = signal.butter(order, f / (fs/2))
    return lfilter(b, a, s)

def notchfilter(s, f, fs, Q = 12,):
    b, a = signal.iirnotch(f / (fs/2), Q)
    return lfilter(b, a, s)

```

```python
#data = loadtxt("pruba_medida.txt")
data = loadtxt("ojos_cerrados_y_blinks_aluminio.txt")
fs=100
x = data[:, 5]
#tline = np.linspace(0, len(x), len(x)/fs)

xpa = highpass(x, 0.5, fs)
xpb = lowpass(xpa, 49, fs)
xf = notchfilter(xpb, 50, fs)
f, time, Zxx = signal.stft(xf, fs, nperseg=128)
shape(Zxx[0:50])
pcolormesh(time, f[0:50], np.abs(Zxx)[0:50], vmin=0, vmax=60)
```

```python
E = real(Zxx*conj(Zxx))
pcolormesh(time, f[0:50], E[0:50], vmin=0, vmax=5000)
```

```python
Et = sum(E, 0)
plot(time, Et)
```

```python
def frec2sample_range(fi, fo, fs, N):
    si = max(1,floor(fi*N/fs))
    so = ceil(fo*N/fs)
    return int(si), int(so)

def band_energy(fft, fs):
    N = len(fft)
    # Energia total
    et = sum(fft)
    dsi, dso = frec2sample_range(0.5, 4, fs, N)
    d = sum(fft[dsi:dso])/et
    tsi, tso = frec2sample_range(4, 7, fs, N)
    t = sum(fft[tsi:tso])/et
    asi, aso = frec2sample_range(7, 13, fs, N)
    a = sum(fft[asi:aso])/et
    bsi, bso = frec2sample_range(13, 30, fs, N)
    b = sum(fft[bsi:bso])/et
    gsi, gso = frec2sample_range(30, 50, fs, N)
    g = sum(fft[gsi:gso])/et
    return d, t, a, b, g
    
```

```python
d, t, a, b, g = band_energy(E[:, 1], fs)
```

```python
bands_time_matrix = zeros((5, shape(E)[1]))
len(E[1,:])
```

```python
nventanas = len(E[1,:])
for i in range(0, nventanas):
    bands_time_matrix[:, i] = transpose(band_energy(E[:, i], fs))
```

```python
plot(time, bands_time_matrix[3, :])
```

```python
pcolormesh(time, f[0:100], np.abs(Zxx)[0:100], vmin=0, vmax=60)
```

```python
for i in range(0, nventanas):
    bands_smoothed = lowpass(bands_time_matrix, 3, 200, order = 5)
```

```python
plot(time, bands_smoothed[0,:], label = 'delta')
plot(time, bands_smoothed[1,:], label = 'theta')
plot(time, bands_smoothed[2,:], label = 'alpha')
plot(time, bands_smoothed[3,:], label = 'beta')
plot(time, bands_smoothed[4,:], label = 'gamma')
legend()
```
