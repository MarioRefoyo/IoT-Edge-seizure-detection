```python
def frec2sample_range(fi, fo, fs, N):
    si = max(1,floor(fi*N/fs))
    so = ceil(fo*N/fs)
    return int(si), int(so)

def band_energy(fft, fs):
    N = len(fft)
    # Energia total
    et = sum(fft)
    dsi, dso = frec2sample_range(0.5, 4, fs, N)
    d = sum(fft[dsi:dso])
    tsi, tso = frec2sample_range(4, 7, fs, N)
    t = sum(fft[tsi:tso])
    asi, aso = frec2sample_range(7, 13, fs, N)
    a = sum(fft[asi:aso])
    bsi, bso = frec2sample_range(13, 30, fs, N)
    b = sum(fft[bsi:bso])
    gsi, gso = frec2sample_range(30, 50, fs, N)
    g = sum(fft[gsi:gso])
    return et, d, t, a, b, g
```

```python
#a = slicedf.loc[1, slicedf['sample1':'sample512']]
ninstances = df.shape[0]
fs = 256

et = zeros(ninstances)
d = zeros(ninstances)
t = zeros(ninstances)
a = zeros(ninstances)
b = zeros(ninstances)
g = zeros(ninstances)

for index, row in df.iterrows():
    signal = row[0:512].values
    X = fft(signal)
    Px = real(X*conj(X))
    et[index], d[index], t[index], a[index], b[index], g[index] = band_energy(Px, fs)
    
```

```python
slicedf['total_energy'] = et
slicedf['delta'] = pd.Series( d, index = slicedf.index)
slicedf['theta'] = pd.Series( t, index = slicedf.index)
slicedf['alpha'] = pd.Series( a, index = slicedf.index)
slicedf['alpha'] = pd.Series( a, index = slicedf.index)
slicedf['gamma'] = pd.Series( g, index = slicedf.index)
```
