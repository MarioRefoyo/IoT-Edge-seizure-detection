```python
timeW = 2
N = timeW*fs
nw = int(signals.shape[1]/N)
```

```python
ch1 = np.reshape(signals[0,:], [nw, N])
ch1.shape
```

```python
def frec2sample_range(fi, fo, fs, N):
    si = max(1,floor(fi*N/fs))
    so = ceil(fo*N/fs)
    return int(si), int(so)

def band_energy(fft, fs):
    N = len(fft)
    # Energia total
    et = sum(fft)
    dsi, dso = frec2sample_range(0.5, 4, fs, N)
    d = sum(fft[dsi:dso])
    tsi, tso = frec2sample_range(4, 7, fs, N)
    t = sum(fft[tsi:tso])
    asi, aso = frec2sample_range(7, 13, fs, N)
    a = sum(fft[asi:aso])
    bsi, bso = frec2sample_range(13, 30, fs, N)
    b = sum(fft[bsi:bso])
    gsi, gso = frec2sample_range(30, 50, fs, N)
    g = sum(fft[gsi:gso])
    return et, d, t, a, b, g
```

```python
f, timeAxis, Zxx = signal.stft(signals[0,:], fs, nperseg=N-1)

E = real(Zxx*conj(Zxx))
pcolormesh(timeAxis, f, np.abs(Zxx), vmin=0, vmax=30)
```

```python
et = zeros(nw)
d = zeros(nw)
t = zeros(nw)
a = zeros(nw)
b = zeros(nw)
g = zeros(nw)

spectral_centroid = zeros(nw)
variatonial_coeff = zeros(nw)
spectral_skew = zeros(nw)

for n, s in enumerate(ch1):
    X = fft(s)
    Px = real(X*conj(X))
    et[n], d[n], t[n], a[n], b[n], g[n] = band_energy(Px, fs)
    p_spectrum = Px[:N//2]/sum(Px[:N//2])
    normalized_f = linspace(0, 1, N//2)
    spectral_centroid[n] = sum(normalized_f * p_spectrum)
    variatonial_coeff[n] = sum( (((normalized_f - spectral_centroid[n]))**2) * p_spectrum) / sum(p_spectrum)
    spectral_skew[n] = sum( ((normalized_f - spectral_centroid[n])/variatonial_coeff[n])**3 * p_spectrum)

#Normalise
et = (et)/std(et)
d = (d)/std(d)
t = (t)/std(t)
a = (a)/std(a)
b = (b)/std(b)
g = (g)/std(g)



variatonial_coeff = variatonial_coeff/max(variatonial_coeff)
spectral_skew = spectral_skew/max(spectral_skew)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt
```

```python
def exponential_smooth(timeseries, alpha):
    model = SimpleExpSmoothing(timeseries)
    fit = model.fit(smoothing_level=alpha)
    return fit.fittedvalues
```

```python
import pandas as pd
```

```python
def exponential_smooth(timeseries):
    model = SimpleExpSmoothing(timeseries)
    fit = model.fit(smoothing_level=.3)
    return fit.fittedvalues
```

```python
seizures = {
            'total_energy': exponential_smooth(et),
            'delta': exponential_smooth(d),
            'theta': exponential_smooth(t),
            'alpha': exponential_smooth(a),
            'beta': exponential_smooth(b),
            'gamma': exponential_smooth(g),
            'spectral_centroid': spectral_centroid,
            'variatonial_coeff': variatonial_coeff,
            'spectral_skew': spectral_skew,
            'seizure': seizure
            }

df = pd.DataFrame(seizures, columns = ['total_energy', 'delta', 'theta', 'alpha', 'beta', 'gamma', 'spectral_centroid', 'variatonial_coeff', 'spectral_skew', 'seizure'])
df["seizure"] = df["seizure"].astype(bool)
```
