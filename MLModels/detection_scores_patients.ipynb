```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
knn_data = pd.read_csv('results_detection\\{}_1channels.csv'.format('knn'), delimiter=',')
rnf_data = pd.read_csv('results_detection\\{}_1channels.csv'.format('rnf'), delimiter=',')
svm_data = pd.read_csv('results_detection\\{}_1channels.csv'.format('svm_rbf'), delimiter=',')
all_data = pd.concat([knn_data, rnf_data, svm_data]).reset_index(drop=True)
```

```python
all_data
```

```python
sns.set(style="white", font_scale=1.5)
g = sns.catplot(x="patient", y="roc_auc", hue="model", data=all_data, height = 10, aspect=2, kind="bar", palette="muted")
g.fig.suptitle('Roc area of every patient', fontsize=38)
g.set(ylim=(0.49, 1))
```

```python
sns.set(style="white", font_scale=1.5)
g = sns.catplot(x="patient", y="sensitivity", hue="model", data=all_data, height = 10, aspect=2, kind="bar", palette="muted")
g.fig.suptitle('Sensitivity of every patient', fontsize=30)
```

```python
sns.set(style="white", font_scale=1.5)
g = sns.catplot(x="patient", y="specificity", hue="model", data=all_data, height = 10, aspect=2, kind="bar", palette="muted")
g.despine(left=True)
g.fig.suptitle('Specificity of every patient', fontsize=30)
```

```python
sns.set(style="white", font_scale=1.5)
g = sns.catplot(x="patient", y="precision", hue="model", data=all_data, height = 10, aspect=2, kind="bar", palette="muted")
g.despine(left=True)
g.fig.suptitle('Precision of every patient', fontsize=30)
```
