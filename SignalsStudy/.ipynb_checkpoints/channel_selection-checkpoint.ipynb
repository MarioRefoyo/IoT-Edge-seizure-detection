```python
import pyedflib
import numpy as np
from pylab import *
from scipy import signal
import pandas as pd
import glob, os
import mne
from collections import OrderedDict
from scipy.stats import *


# Select the patient
patient = '\chb01'

basedir = os.getcwd()
dbdir = "E:\Master\TFM\chb-mit-scalp-eeg-database-1.0.0" + patient
```

```python
class Register:
    
    def __init__(self, name, fs, nseizures):
        self.name = name
        self.fs = fs
        self.nseizures = nseizures
        self.seizures = []
        self.channels = []
            
    def addSeizure (self, start, end):
        seizure = [start, end]
        self.seizures.append(seizure)
        
    def addCh (self, channels):
        self.channels = channels
```

# Read the annotations file
## Saved info about all registers in 'registers' dictionary

```python
os.chdir(dbdir)
annotation = glob.glob('*txt')

with open(annotation[0]) as f:
    registers = {}
    channels = []
    for line in f:
        if ("Data Sampling Rate" in line):
            line = line.split()
            fs = int(line[3])            
        if ("Channel " in line):
            line = line.split()
            channels.append(line[2])
        elif ("File Name" in line):
            name = line.split()[2]
            while True:
                newLine = f.readline()
                if ("Number of Seizures" in newLine):
                    nseizures = int(newLine.split()[5])
                    register = Register(name, fs, nseizures)
                    if nseizures > 0:
                        for i in range(nseizures):
                            line1 = f.readline().split()
                            line2 = f.readline().split()
                            if (line1[3] == "Time:"):
                                start = int(line1[4])
                                end = int(line2[4])
                            else:
                                start = int(line1[3])
                                end = int(line2[3])
                            register.addSeizure(start, end)
                            
                    register.addCh(list(OrderedDict.fromkeys(channels)))
                    registers[name] = register
                    break
```

## Read the .edf

```python
def read_data(filename, channels=[]):
    f = pyedflib.EdfReader(filename)
    if len(channels) == 0:
        channels = f.getSignalLabels()

    channel_names = f.getSignalLabels()
    fs = f.getSampleFrequencies()

    data = np.zeros((len(channels), f.getNSamples()[0]))
    for i, channel in enumerate(channels):
        data[i, :] = f.readSignal(channel_names.index(channel))
        
    time = np.linspace(0, data.shape[1]/fs[0], data.shape[1])
    f._close()
    return data, fs[0], time
```

```python
register = registers['chb01_03.edf']
signals, fs, time = read_data(dbdir + '\\' + register.name, register.channels)
```

```python
plot_kwargs_signal = {
    'scalings': dict(eeg=30e-5),   # zooms the plot out
    'highpass': 0.5,              # filters out low frequencies
    'lowpass': 70.,                # filters out high frequencies
    'show_scrollbars': False,
}

def mne_object(data, freq, channels, events = None):
    # create an mne info file with meta data about the EEG
    info = mne.create_info(ch_names=channels, sfreq=freq, ch_types=['eeg']*data.shape[0])
    data = data*1e-6
    raw = mne.io.RawArray(data, info)
    
    if events:
        matrix = np.array(events)
        start_times = matrix[:, 0]
        end_times = matrix[:, 1]
        anno_length = end_times-start_times
        event_name = np.array(['Ictal']*len(anno_length))

        raw.set_annotations(mne.Annotations(onset=start_times, duration=anno_length, description=event_name))

    return raw
```

```python
mne_data = mne_object(signals, register.fs, register.channels, register.seizures)
```

# Calculate variance and entrophy

### Slice the signal in windows

```python
def windowing(data, timeW, fs):
    samples = data.shape[1]
    timeW = 2
    N = timeW*fs
    nw = int(samples//N)

    data = data [:, 0:nw*N]
    time = np.linspace(0, signals.shape[1]/fs, signals.shape[1])
    return data, time, nw, N
```

```python
timeW = 2
signals, time, nw, N = windowing(signals, timeW, fs)
```

### Create the window seizure vector

```python
seizure = zeros(signals.shape[1])
for n in range (len(register.seizures)):
    start = register.seizures[n][0]*fs
    end = register.seizures[n][1]*fs
    seizure[start:end] = np.ones(end-start)

seizureW = np.reshape(seizure, [nw, N])
seizureW = (sum(seizureW, 1) > N//2)
plot(seizureW)
```

### Calculate variance and entropy

```python
variances = np.zeros([signals.shape[0], nw])
entropys = np.zeros([signals.shape[0], nw])
for channel, signal in enumerate(signals):
    signal = np.reshape(signal, [nw, N])
    variances[channel, :] = np.var(signal, 1)
    for n, s in enumerate(signal):
        pd_series = pd.Series(s)
        counts = pd_series.value_counts()
        entropys[channel, n] = entropy(counts, base = 2)
        
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt
```

```python
def exponential_smooth(timeseries, alpha):
    model = SimpleExpSmoothing(timeseries)
    fit = model.fit(smoothing_level=alpha)
    return fit.fittedvalues
```

```python
s_entropys = zeros(entropys.shape)
for i in range(entropys.shape[0]):
    s_entropys[i, :] = exponential_smooth(entropys[i, :], 0.2)

plot(s_entropys[0,:])
```

### Plot results

```python
plot_kwargs_var = {
    'scalings': dict(eeg=80e-4),   # zooms the plot out
    'show_scrollbars': False,
}

plot_kwargs_ent = {
    'scalings': dict(eeg=80e-8),   # zooms the plot out
    'show_scrollbars': False,
}
```

```python
mne_var = mne_object(variances, 1/timeW, register.channels, register.seizures)
mne_ent = mne_object(entropys, 1/timeW, register.channels, register.seizures)
```

```python
mne_var.plot(start = register.seizures[0][0]- 100, duration = 250, **plot_kwargs_var)
mne_ent.plot(start = register.seizures[0][0]- 100, duration = 250, **plot_kwargs_ent)
mne_data.plot(start = register.seizures[0][0]- 100, duration = 250, **plot_kwargs_signal)
```
